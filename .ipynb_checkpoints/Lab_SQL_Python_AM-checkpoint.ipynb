2. Write a Python function called rentals_month that retrieves rental data for a given month and year (passed as parameters) from the Sakila database as a Pandas DataFrame. The function should take in three parameters:

- engine: an object representing the database connection engine to be used to establish a connection to the Sakila database.
- month: an integer representing the month for which rental data is to be retrieved.
- year: an integer representing the year for which rental data is to be retrieved.

The function should execute a SQL query to retrieve the rental data for the specified month and year from the rental table in the Sakila database, and return it as a pandas DataFrame.

```python
def rental_per_month(engine, month, year):
    month_input = str(month) #this declares the month as string so I can combine later on in the query
    year_input = str(year) #same for year
    query = str('''SELECT * FROM sakila.rental WHERE right(left(rental_date,7),2) = '''+month_input+''' AND left(rental_date,4) = '''+year_input)
    output = pd.read_sql_query(query, engine) #requesting the data via engine
    return output
```

```python
rentals_may = rental_per_month(engine, 5, 2005)
rentals_may #1156 rows, which is the same when computed in SQL workbench
```

3. Develop a Python function called rental_count_month that takes the DataFrame provided by rentals_month as input along with the month and year and returns a new DataFrame containing the number of rentals made by each customer_id during the selected month and year.

The function should also include the month and year as parameters and use them to name the new column according to the month and year, for example, if the input month is 05 and the year is 2005, the column name should be "rentals_05_2005".

```python
def rental_count_month(rental_df, month, year):
    output = rental_df.groupby(['customer_id']).count() #grouping by customer_id
    output = output.reset_index() #reseting the index so that 'customer_id' could be added as a new column
    month_input = str(month) #declaring as string so I can use as input in new column
    year_input = str(year) #same for year
    ref_date = str('rentals_0'+month_input+'_'+year_input) #creating this as input to new column
    output['ref_date'] = ref_date #adding new column with reference date
    final_table = output[['customer_id', 'rental_id', 'ref_date']].copy() #selecting only the columns I want
    return final_table #returns the count of 'rental_id' (rentals) by 'customer_id' and new column 'ref_date'
```

```python
#Using the rentals_made pd from 2.:
rental_count_may = rental_count_month(rentals_may, 5, 2005)
rental_count_may
```

4. Create a Python function called compare_rentals that takes two DataFrames as input containing the number of rentals made by each customer in different months and years. The function should return a combined DataFrame with a new 'difference' column, which is the difference between the number of rentals in the two months.

```python
# Creating rentals_june and rental_count_june using the functions from above so I can use to test function later on:
rentals_june = rental_per_month(engine, 6, 2005)
rentals_june
```

```python
rental_count_june = rental_count_month(rentals_june, 6, 2005)
rental_count_june
```

```python
#checking if same lenght, that is if the same number of customers are listed in both months
display(rental_count_may['customer_id'].value_counts())
display(rental_count_june['customer_id'].value_counts())
#as there could be difference between customer_id between the two months, there could be NaNs after merging
data2 = pd.merge(rental_count_may, rental_count_june, how='outer', on='customer_id') #merged using outer so it includes all rows and not only matching ones
display(data2.isna().sum()) 
#indeed, will need to fill with 0 whenever 'rental_id' is missing, as it implies that customer did not have rentals that month
data2['rental_id_x'] = data2['rental_id_x'].fillna(0)
data2['rental_id_y'] = data2['rental_id_y'].fillna(0)
#will do the same with 'rental_reference' using first row as reference
data2['ref_date_x'] = data2['ref_date_x'].fillna(data2.loc[0,'ref_date_x'])
data2['ref_date_y'] = data2['ref_date_y'].fillna(data2.loc[0,'ref_date_y'])
#now adding a new column with column as difference (y-x):
data2['difference (y-x)'] = data2['rental_id_y']-data2['rental_id_x']
data2
#will add all of these steps in a function below
```

```python
def compare_rentals(rental_count1, rental_count2):
    data2 = pd.merge(rental_count1, rental_count2, how='outer', on='customer_id')
    data2['rental_id_x'] = data2['rental_id_x'].fillna(0)
    data2['rental_id_y'] = data2['rental_id_y'].fillna(0)
    data2['ref_date_x'] = data2['ref_date_x'].fillna(data2.loc[0,'ref_date_x'])
    data2['ref_date_y'] = data2['ref_date_y'].fillna(data2.loc[0,'ref_date_y'])
    data2['difference (y-x)'] = data2['rental_id_y']-data2['rental_id_x']
    return data2
```

```python
#now testing the function using the dataframes rental_count_may and rental_count_june
compare_rentals(rental_count_may, rental_count_june)
```
